```python
#Importwoanie potrzebnych bibliotek (losowość i tworzenie wykresów)
from random import random
import matplotlib.pyplot as plt

n0 = 1000    #początkowa ilość jąder pierwszego pierwiastka
dt = 1      #krok czasowy w sekundach
dur = 10000 #czas trwania w sekundach
```

```python
#Definicja funkcji f opisanej w opisie algorytmu
def f(s):
  if(s=='PO218'):
      return ['PB214',3.05*60]
  elif(s=='PB214'):
      return ['BI214',26.8*60]
  elif(s=='BI214'):
      return ['PB210',19.7*60]
  elif(s=='PB210'):
      return ['PB210',1]
  elif(s=='PB211'):
      return ['BI211',36.1*60]
  elif(s=='BI211'):
      return ['TL207',2.14*60]
  elif(s=='TL207'):
      return ['PB207',4.77*60]
  elif(s=='PB207'):
      return ['PB207',1]
```

```python
#Ustalenie rozmiarów wykresu
plt.rcParams["figure.figsize"] = (20,10)

#Pętla zapewniająca 5 symulacji
for superi in range(5):
  tab = ['PO218'] * n0

  #Tablica do pamiętania ilości danych jąder w danym momencie
  x = [[0 for a in range(int(dur/dt))] for b in range(5)] 

  for i in range(int(dur/dt)):
    for j in range(len(tab)):
      #Symulacja rozpadu jąder
      if(random() < (1 - (1/2)**(dt / (f(tab[j])[1])))):
          tab[j] = f(tab[j])[0]
    #Zapisanie liczby wystąpień danych jąder
    x[0][i] = tab.count('PO218')
    x[1][i] = tab.count('PB214')
    x[2][i] = tab.count('BI214')
    x[3][i] = tab.count('PO214')
    x[4][i] = tab.count('PB210')
      
  #Stworzenie wykresu
  plt.plot(x[0],color='red')
  plt.plot(x[1],color='blue')
  plt.plot(x[2],color='green')
  plt.plot(x[3],color='yellow')
  plt.plot(x[4],color='cyan')
  plt.title("Symulacja szeregu uranowo-radowego")
  plt.ylabel("Ilość jąder")
  plt.xlabel("Ilość kroków czasowych")

#Wypisanie i wyczyszczenie wykresu
plt.show()
plt.clf()
plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
```
